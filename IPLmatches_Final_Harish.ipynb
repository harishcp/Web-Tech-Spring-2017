```python
from sklearn.preprocessing import LabelEncoder                                    # Label encoder is used to transform non-numerical labels to numerical labels.
number = LabelEncoder()                                                           # Here as our data set consists only of categorical values except age
for i in iplmatches_new.columns:                                                          # in order to find the correlation between variables, we do Label encoding 
    iplmatches_new[i] = number.fit_transform(iplmatches_new[i].astype('str'))
iplmatches_new.dtypes
```

```python
corr=iplmatches_new.corr()
corr
```

```python
tosswinner_winner = pd.DataFrame({'Toss Winner': iplmatches_new['toss_winner'], 'winner': iplmatches_new['winner']}, dtype='category') # Create a new dataframe with the two columns and assign numbers in place of their categories
df_num = tosswinner_winner.apply(lambda x: x.cat.codes)
tosswinner_winner
df_num
print(df_num.corr())             # Run a correlation calculation 
```

```python
#To analyse the corelation, we do label encoding since the columns have max categorical columns
import seaborn as sns
features = df_num.corr()
plt.figure(figsize=(4,4))
sns.heatmap(features,vmax=1,square=True,annot=False,cmap='Blues')
plt.show()
```
